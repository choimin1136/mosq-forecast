```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/16~25 강수 정보(공란 강수x).csv', encoding="cp949", sep=",") #파일 불러오기 header사용하지 않고 파일에서 설명문을 직접 지움.
df2 = pd.read_csv('/content/16~25 기온 정보.csv', encoding="cp949", sep=",")
df3 = pd.read_csv('/content/16~25 습도 정보.csv', encoding="cp949", sep=",")
df4 = pd.read_csv('/content/16~25 일조 정보.csv', encoding="cp949", sep=",")
df5 = pd.read_csv('/content/16~25 풍속 정보.csv', encoding="cp949", sep=",")
df6 = pd.read_csv('/content/서울시 모기예보제 정보.csv', encoding="cp949", sep=",")
```

```python
print(df2.columns)
```

```python
#날짜 생성
df1['date'] = pd.to_datetime(df1['일시'])
df2['date'] = pd.to_datetime(df2['일시'])
df3['date'] = pd.to_datetime(df3['일시'])
df4['date'] = pd.to_datetime(df4['일시'])
df5['date'] = pd.to_datetime(df5['일시'])
df6['date'] = pd.to_datetime(df6['모기지수 발생일'])
```

```python
#필요한 컬럼 선택
df1 = df1[['date','강수량(mm)', '1시간최다강수량(mm)', '1시간최다강수량시각']]
df2 = df2[['date','평균기온(℃)', '최고기온(℃)', '\t최고기온시각', '최저기온(℃)','최저기온시각일교차' ]]
df3 = df3[['date','평균습도(%rh)', '최저습도(%rh)']]
df4 = df4[['date','일조합(hr)', '일조율(%)', '일사합(MJ/m2)']]
df5 = df5[['date','평균풍속(m/s)', '최대풍속(m/s)', '최대풍속풍향(deg)', '\t최대풍속시각', '최대순간풍속(m/s)', '최대순간풍속풍향(deg)', '\t최대순간풍속시각']]
df6 = df6[['date','모기지수(주거지)', '모기지수(공원)']]
```

```python
#merge
df = df1.merge(df2, on='date', how='outer') \
        .merge(df3, on='date', how='outer') \
        .merge(df4, on='date', how='outer') \
        .merge(df5, on='date', how='outer') \
        .merge(df6, on='date', how='outer')

#정렬
df = df.sort_values('date')
#2020년 이후 데이터만 사용
df = df[df['date'] >= '2020-01-01']

print(df.head())
```
